```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset load
data = pd.read_csv(r"C:\Users\revan\Downloads\restaurant_customer_satisfaction.csv")
```

```python
# Average Spend by Gender
avg_spend_by_gender = data.groupby('Gender')['AverageSpend'].mean()
print(avg_spend_by_gender)
```

```python
# Visualizing Average Spend by Gender
plt.figure(figsize=(8,6))
sns.barplot(x=['Female', 'Male'], y=avg_spend_by_gender.values)
plt.title('Average Spend by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Spend')
plt.show()
```

```python
# Distribution of customer ages
plt.figure(figsize=(10,6))
sns.histplot(data['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation between Wait Time and Service Rating, Food Rating, Ambiance Rating
correlation = data[['WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']].corr()
print(correlation)
```

```python
# Heatmap of correlations
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Satisfaction Analysis: Correlation of HighSatisfaction with other factors
satisfaction_corr = data[['HighSatisfaction', 'WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']].corr()
print(satisfaction_corr)
```

```python
# Bar plot of satisfaction levels by gender
plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='HighSatisfaction', data=data)
plt.title('High Satisfaction by Gender')
plt.xlabel('Gender')
plt.ylabel('High Satisfaction')
plt.show()
```

```python
# Scatter plot of Wait Time vs Satisfaction
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='WaitTime', y='HighSatisfaction', hue='Gender')
plt.title('Wait Time vs High Satisfaction')
plt.xlabel('Wait Time')
plt.ylabel('High Satisfaction')
plt.show()
```

```python
# Box plot of Service Rating by Dining Occasion
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='DiningOccasion', y='ServiceRating')
plt.title('Service Rating by Dining Occasion')
plt.show()
```

```python
# Convert DataFrame to numpy array for analysis
data_np = data.to_numpy()
```

```python
# Basic Statistics
print("Basic Statistics:")
print(f"Mean Age: {np.mean(data_np[:, 1])}")
print(f"Mean Income: {np.mean(data_np[:, 3])}")
print(f"Mean Average Spend: {np.mean(data_np[:, 5])}")
print(f"Mean Wait Time: {np.mean(data_np[:, 14])}")
```

```python
# Median values
print("\nMedian Values:")
print(f"Median Age: {np.median(data_np[:, 1])}")
print(f"Median Income: {np.median(data_np[:, 3])}")
print(f"Median Average Spend: {np.median(data_np[:, 5])}")
print(f"Median Wait Time: {np.median(data_np[:, 14])}")
```

```python
# Standard Deviation
print("\nStandard Deviations:")
print(f"Standard Deviation of Age: {np.std(data_np[:, 1])}")
print(f"Standard Deviation of Income: {np.std(data_np[:, 3])}")
print(f"Standard Deviation of Average Spend: {np.std(data_np[:, 5])}")
print(f"Standard Deviation of Wait Time: {np.std(data_np[:, 14])}")
```

```python
# High Satisfaction Rate
high_satisfaction_rate = np.mean(data_np[:, 17])
print(f"\nHigh Satisfaction Rate: {high_satisfaction_rate:.2%}")
```
