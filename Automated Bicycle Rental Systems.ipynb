```python
# Summary statistics
print(data.describe())
```

```python
# Scatter plot of temperature vs rental count
plt.figure(figsize=(8, 6))
sns.scatterplot(x='temp', y='cnt', data=data)
plt.title('Temperature vs Rental Count')
plt.xlabel('Temperature')
plt.ylabel('Rental Count')
plt.show()
```

```python
# Line plot of count over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='dteday', y='cnt', data=data)
plt.title('Rental Count Over Time')
plt.xlabel('Date')
plt.ylabel('Rental Count')
plt.show()
```

```python
# Group by weekday
weekday_group = data.groupby('weekday')['cnt'].mean()
```

```python
# Bar plot of average rentals by weekday
plt.figure(figsize=(8, 6))
sns.barplot(x=weekday_group.index, y=weekday_group.values)
plt.title('Average Rentals by Weekday')
plt.xlabel('Weekday (0 = Sunday)')
plt.ylabel('Average Rental Count')
plt.show()
```

```python
# Distribution of 'temp' (Temperature)
sns.histplot(data['temp'], kde=True)
plt.title("Distribution of Temperature")
plt.show()
```

```python
# Scatter plot for temperature vs total count
sns.scatterplot(x='temp', y='cnt', data=data)
plt.title("Temperature vs Total Count")
plt.show()
```

```python
# Pairplot to explore relationships between multiple variables
sns.pairplot(data[['temp', 'hum', 'windspeed', 'cnt']])
plt.show()
```
