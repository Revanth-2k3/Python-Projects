```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset load
data = pd.read_csv(r"C:\Users\revan\Downloads\crime_data.csv")
```

```python
crime_by_area = data['AREA NAME'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=crime_by_area.index, y=crime_by_area.values)
plt.title('Crime Count by Area')
plt.xticks(rotation=90)
plt.ylabel('Number of Crimes')
plt.show()
```

```python
crime_by_type = data['Crm Cd Desc'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=crime_by_type.values, y=crime_by_type.index)
plt.title('Crime Count by Type')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Analyze crimes by weapon type
weapon_usage = data['Weapon Desc'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=weapon_usage.index, y=weapon_usage.values)
plt.title('Crimes by Weapon Type')
plt.xticks(rotation=90)
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Victim demographics - Age and Sex
plt.figure(figsize=(10,6))
sns.histplot(data['Vict Age'], bins=20)
plt.title('Victim Age Distribution')
plt.show()
```

```python
# Crime distribution by victim sex
vict_sex_dist = data['Vict Sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(vict_sex_dist, labels=vict_sex_dist.index, autopct='%1.1f%%')
plt.title('Victim Sex Distribution')
plt.show()
```

```python
# Crimes over time (Date Rptd)
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'])
crime_over_time = data.groupby(data['DATE OCC'].dt.year)['Crm Cd'].count()

plt.figure(figsize=(10,6))
sns.lineplot(x=crime_over_time.index, y=crime_over_time.values)
plt.title('Crimes Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Vict Age'], bins=range(0, 70, 10), edgecolor='black')
plt.title('Histogram of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
area_counts = data['AREA'].value_counts()

plt.figure(figsize=(12, 8))
area_counts.plot(kind='bar')
plt.title('Number of Crimes in Different Areas')
plt.xlabel('Area')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
sns.histplot(data['Vict Age'], bins=10, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
