```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset load
data = pd.read_csv(r"C:\Users\revan\Downloads\powerlifting_dataset.csv")
```

```python
# Grouping by lift type and summarizing the amount lifted
lift_summary = data.groupby('Lift Type')['Amount Lifted (kg)'].describe()
print(lift_summary)
```

```python
# Average amount lifted by weight class
avg_lift_by_weight_class = data.groupby('Weight Class')['Amount Lifted (kg)'].mean()
print(avg_lift_by_weight_class)
```

```python
# Visualizing the average amount lifted by weight class
plt.figure(figsize=(10,6))
sns.barplot(x=avg_lift_by_weight_class.index, y=avg_lift_by_weight_class.values)
plt.title('Average Amount Lifted by Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Average Amount Lifted (kg)')
plt.show()
```

```python
# Visualizing lift type distribution
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='Lift Type')
plt.title('Distribution of Lift Types')
plt.xlabel('Lift Type')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of Age vs Amount Lifted, categorized by Lift Type
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Age', y='Amount Lifted (kg)', hue='Lift Type')
plt.title('Age vs Amount Lifted by Lift Type')
plt.show()
```

```python
# Distribution of Amount Lifted
plt.figure(figsize=(8,6))
sns.histplot(data['Amount Lifted (kg)'], bins=10)
plt.title('Distribution of Amount Lifted')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Grouping by lifter name to get the total amount lifted by each individual
total_lift_by_lifter = data.groupby('Lifter Name')['Amount Lifted (kg)'].sum()
print(total_lift_by_lifter)
```

```python
# Visualizing total amount lifted by each lifter
plt.figure(figsize=(10,6))
sns.barplot(x=total_lift_by_lifter.index, y=total_lift_by_lifter.values)
plt.title('Total Amount Lifted by Each Lifter')
plt.xticks(rotation=90)
plt.ylabel('Total Amount Lifted (kg)')
plt.show()
```

```python
# Distribution of Amount Lifted
plt.figure(figsize=(10, 6))
plt.hist(data['Amount Lifted (kg)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Amount Lifted')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Distribution of Amount Lifted
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount Lifted (kg)'], bins=40, color='skyblue', kde=True)
plt.title('Distribution of Amount Lifted')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the occurrences of each lift type
lift_counts = data['Lift Type'].value_counts()

# Create the pie chart
plt.pie(lift_counts, labels=lift_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Lift Types')

# Display the chart
plt.show()
```
