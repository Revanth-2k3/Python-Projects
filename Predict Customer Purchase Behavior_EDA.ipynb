```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x='Gender', y='AnnualIncome', data=data)
plt.title("Annual Income Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(11, 5))
sns.countplot(x='PurchaseStatus', data=data)
plt.title("Purchase Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='PurchaseStatus', y='TimeSpentOnWebsite', data=data)
plt.title("Time Spent on Website vs. Purchase Status")
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.countplot(x='LoyaltyProgram', hue='PurchaseStatus', data=data)
plt.title("Loyalty Program vs Purchase Status")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='PurchaseStatus', y='DiscountsAvailed', data=data)
plt.title("Discounts Availed vs Purchase Status")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data,x='Age',bins=30,kde=True)
```

```python
correlation_income_purchases = data[['AnnualIncome', 'NumberOfPurchases']].corr().iloc[0, 1]
print(f"\nCorrelation between Annual Income and Number of Purchases: {correlation_income_purchases:.2f}")
```

```python
# Group by Purchase Status and calculate mean values
grouped = data.groupby('PurchaseStatus').mean()
print("\nMean Values by Purchase Status:")
print(grouped)
```

```python
# Additional Analysis: Count of PurchaseStatus
status_counts = data['PurchaseStatus'].value_counts()
print("\nCount of Purchase Status:")
print(status_counts)
```

```python
plt.subplot(3, 2, 2)
plt.hist(data['AnnualIncome'], bins=10, color='salmon', edgecolor='black')
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
```

```python
avg_purchases_by_status = data.groupby('PurchaseStatus')['NumberOfPurchases'].mean()
plt.figure(figsize=(8, 6))
avg_purchases_by_status.plot(kind='bar', color='teal')
plt.title('Average Number of Purchases by Purchase Status')
plt.xlabel('Purchase Status')
plt.ylabel('Average Number of Purchases')
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot distributions using seaborn
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(data['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(data['AnnualIncome'], bins=10, kde=True, color='salmon')
plt.title('Annual Income Distribution')
```

```python
# Scatter plot for Annual Income vs Number of Purchases
plt.figure(figsize=(8, 6))
sns.scatterplot(x='AnnualIncome', y='NumberOfPurchases', data=data, hue='PurchaseStatus', palette='deep')
plt.title('Annual Income vs Number of Purchases')
plt.xlabel('Annual Income')
plt.ylabel('Number of Purchases')
plt.legend(title='Purchase Status')
plt.grid(True)
plt.show()
```

```python
# Pair plot to see relationships between numerical features
sns.pairplot(data, hue='PurchaseStatus', palette='husl')
plt.show()
```

```python
correlation_matrix = data.corr()
# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count the occurrences of each purchase status
purchase_status_counts = data['PurchaseStatus'].value_counts()

# Create the pie chart
plt.pie(purchase_status_counts, labels=purchase_status_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Purchase Statuses')

# Display the chart
plt.show()
```
