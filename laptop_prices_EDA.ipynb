```python
# Select only the numeric columns
numeric_data = data[['Inches', 'Ram', 'Weight', 'Price_euros']]

# Calculate the mean, median, and standard deviation for each numeric column
mean_values = np.mean(numeric_data, axis=0)
median_values = np.median(numeric_data, axis=0)
std_dev_values = np.std(numeric_data, axis=0)

print("Mean values:")
print(mean_values)

print("Median values:")
print(median_values)

print("Standard deviation values:")
print(std_dev_values)
```

```python
# Find the most expensive laptop
most_expensive_laptop = data.loc[data['Price_euros'].idxmax()]
print(most_expensive_laptop)
```

```python
# Distribution of Laptop Types
plt.figure(figsize=(8, 6))
plt.bar(data['Price_euros'].value_counts().index, data['Price_euros'].value_counts())
plt.title('Distribution of Laptop Types')
plt.xlabel('Laptop Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship Between Screen Size and Price
plt.figure(figsize=(8, 6))
plt.scatter(data['Inches'], data['Price_euros'])
plt.title('Relationship Between Screen Size and Price')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# Count the occurrences of each CPU manufacturer
cpu_manufacturer_counts = data['CPU_company'].value_counts()

# Create the pie chart
plt.pie(cpu_manufacturer_counts, labels=cpu_manufacturer_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of CPU Manufacturers')

# Display the chart
plt.show()
```

```python
# Bar plot to visualize the distribution of laptop types
sns.set_theme()
plt.figure(figsize=(10, 6))
sns.countplot(x='Weight', data=data)
plt.title('Distribution of Laptop Types')
plt.xlabel('Laptop Type')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot to visualize the relationship between price and screen size
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Inches', y='Price_euros', data=data)
plt.title('Relationship Between Screen Size and Price')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# Create a violin plot to show the distribution of prices by laptop type
plt.figure(figsize=(10, 6))
sns.violinplot(x='Inches', y='Price_euros', data=data)
plt.title('Price Distribution by Laptop Type')
plt.xlabel('Laptop Type')
plt.ylabel('Price (Euros)')
```
